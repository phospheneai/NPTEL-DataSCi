```python
import numpy as np
from sklearn import preprocessing, svm 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
selected_cols = ['Area', 'Production']
```

```python
df = pd.read_csv('apy.csv', usecols=selected_cols)
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
len(df)
```

```python
plt.scatter(df['Area'], df['Production'])
```

```python
X = np.array(df['Area']).reshape(-1, 1) 
y = np.array(df['Production']).reshape(-1, 1) 
```

```python
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=111) 
```

```python
lin_reg = LinearRegression() 
```

```python
lin_reg.fit(X_train, y_train)
```

```python
print(lin_reg.score(X_test, y_test)) 
```

```python
preds = lin_reg.predict(X_train)
```

```python
preds.shape, y_train.shape
```

```python
r2_score(y_train, preds)
```
